## Thickness Fitting

This function helps to solve the inverse problem; given a S(E) curve, we seek a best-fit thickness of the oxide layer.
The issue is, that d = f(S). So the code generates a curve based on the thickness layer, compares it to the data, and then refines until an adequate estimate is established. 


It tastes as input the energies (also one value), the thickness of the oxidation layer, the S parameter at the surface $S_{surf}$, the density of the oxide $\rho_{ox}$ and the S_parameter of the bulk $S_{bulk}$

Different parameters explained:

$\xi_0 = \frac{\bar{z}}{\Gamma(1 + 1/m)}$ is a way of calculating the characteristic implantation depth of the positron. It consists of the mean depth $\bar{z}$, which we get from the integration alonside a Gamma function and $m$. 


The probability function: 
$$P(z) = \frac{m z^{m-1}}{z_0^m} \exp\left[ -\left( \frac{z}{z_0} \right)^m \right]
$$ 

From the power law, it is experimentally found, that the relation between the mean implantation depth and the energy of the particle and constant A and n.
$$
\bar{z} = A E^n 
$$

This function is the Makhovian profile. The $\bar{z}$ is a mean depth, so from the Makhovian profile and means of integration we get;
$$
\bar{z} = \int_{0}^{\infty} z \cdot P(z) \, dz
$$
This equation makes sense since the we integrate the depth with a 'density' distribution, from which the average depth of implantation arises. 
The characteristic versus mean implantation depth is not the same, since one describes a mean and emperical parameter, the other one generalises it to the material and particle type. So from the integration we recover $z_0$. Isolation and then you have it. Then for a standard positron m parameter of 2, we get the Gamma function equaling $\sqrt{\pi}/2$. So eventually; 
$$
z_0 = \frac{A}{\rho \Gamma \left(1 + \frac{1}{m}\right)} E^n = \frac{A}{\rho \sqrt{\pi}/2} E^n, [\mu g  \ keV^n \ g^{-1} \ cm^2 ]
$$

We can also derive the median implantation depth, since that one is representative of the distribution. 
$$
z_{1/2} = z_0 \ln(2)^{1/m} = \frac{A_{1/2}{\rho}} E^n
$$
where, $A_{1/2} = \frac{A \ln(2)^{1/m}}{\Gamma \left(1 + \frac{1}{m}\right)}$
